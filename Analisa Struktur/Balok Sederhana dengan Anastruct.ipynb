## Penyelesaian

Pertama kita harus mengimport package anastruct sebagai berikut:

```python
from anastruct import SystemElements
ss = SystemElements()
```

Karena Anastruct menggunakan metode elemen hingga maka kita perlu membagi balok menjadi beberapa elemen supaya beban dapat ditemtakan, untuk lebih jelasnya lihat gambar dibawah.
![](https://raw.githubusercontent.com/riskey95/Jupyter-Notebook/master/Gambar/Soal%201%20elemen.jpg)

Balok dibagi menjadi 3 elemen yang masing-masing node dan elemen mempunyai koordinat sebagai berikut:

Tabel Node

|node|i|j
|:---:|:---:|:---:
|1 |0|0
|2 |2|0
|3 |5|0
|4 |9|0

Tabel Elemen

|Elemen   |start|end
|:------- |:---:|:---:
| 1 |(0,0)|(2,0)
| 2 |(2,0)|(5,0)
| 3 |(5,0)|(9,0)

<br>
Kemudian kita membuat elemen dari balok sederhana tersebut:

```python
#elemen 1
ss.add_element(location=[[0,0],[2,0]])
#elemen 2
ss.add_element(location=[[2,0],[5,0]])
#elemen 3
ss.add_element(location=[[5,0],[9,0]])
```

Setelah itu kita tambahkan tumpuan pada ujung-ujung balok:

```python
#tumpuan sendi pada node 1
ss.add_support_hinged(node_id=1)
#tumpuan roll pada node 4
ss.add_support_roll(node_id=4)
```

Selanjutnya kita masukkan beban titik:

```python
#Beban titik pada node 2 sebesar 10kN
ss.point_load(node_id=2, Fy=-10)
#Beban titik pada node 3 sebesar 15kN
ss.point_load(node_id=3, Fy=-15)
```

Sebelum menjalankan program terlebih dahulu kita harus mengecek struktur yang telah dicoding:

```python
ss.show_structure()
```

Jika tidak ditemukan kesalahan, maka langkah terakhir adalah menjalankan program yang telah dicoding:

```python
#untuk menyelesaikan persamaan struktur
ss.solve()

#untuk menampilkan diagram reaksi
ss.show_reaction_force()
```

```python
#untuk menampilkan diagram gaya geser
ss.show_shear_force()

#untuk menampilkan diagram momen
ss.show_bending_moment()
```

```python
#untuk menampilkan diagram gaya normal
ss.show_axial_force()
```
